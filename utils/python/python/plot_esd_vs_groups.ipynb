````python
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
import plotnine as p9
warnings.resetwarnings()
from gudinfo import RunInfo
info = RunInfo(dir)
data = info.f
````

````python
esd_plot=(p9.ggplot(data=data,
          mapping=p9.aes(x='longgroup', y='esd'))
 + p9.geom_point(size=10)
 + p9.scale_y_log10()
 + p9.theme_bw()
 + p9.ggtitle("Functional groups in the 33+16 configuration")
 + p9.xlab("Functional groups")
 + p9.ylab(r'ESD ($\mu$m)')
 + p9.theme(plot_title  = p9.element_text(size=24))
 + p9.theme(axis_title  = p9.element_text(size=20))
 + p9.theme(axis_text_x = p9.element_text(angle=90,size=16))
 + p9.theme(axis_text_y = p9.element_text(size=16))
)
````

The groups are now in order of maximal ESD.

````python
warnings.filterwarnings('ignore')
figfile=os.path.join(dir,'esd')
esd_plot.save(figfile,
              width=10,height=10,dpi=300)
warnings.resetwarnings()
````
