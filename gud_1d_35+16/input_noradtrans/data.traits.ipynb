```python
PALAT_RE=re.compile(r"""^\s*PALAT\(
(?P<prey>.*?)
,
(?P<pred>.*?)
\)=
(?P<val>.*)
,\s*$""",
                   re.VERBOSE)
```

```python
infile=open(infilename,'r')
for line in infile:
    match=PALAT_RE.match(line)
    if match:
        prey=int(match.group("prey"))
        pred=int(match.group("pred"))
        if prey != 15 and prey != 16 and pred != 15 and pred != 16:
            if prey >= 17:
                prey=prey-2
            if pred >= 17:
                pred=pred-2
            lineout=" PALAT({0:3},{1:2})={2},\n".format(
                prey,
                pred,
                match.group("val"))
            outfile.write(lineout)
infile.close()
```
