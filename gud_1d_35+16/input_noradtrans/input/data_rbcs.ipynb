# Generate 1D_BB_NO3_GE_spring.delZ1016mm.32bits.times.*.bin

```python
indir='../../../output/run_20220721_0000_EXP1_0_defineGUD_READ_PARUICE_grp_Xmin1Em2/'
```

```python
import matplotlib
import matplotlib.pyplot as plt
import netcdf_tools
import numpy as np
import os
import os.path
```

```python
infile_no3='1D_BB_NO3_GE_spring.delZ1016mm.32bits.bin'
in_no3=np.fromfile(infile_no3,dtype='>f')
prefix='1D_BB_NO3_GE_spring.delZ1016mm.32bits.times.'
ext='.bin'
```

```python
for f in [0.25,0.50,0.75,1.00,1.25,1.50,1.75,2.00,4.00]:
    out_no3=in_no3*f
    outfile=prefix+'{0:.2f}'.format(f)+ext
    out_no3.astype('>f4').tofile(outfile)
```

## Verification

```python
gridfile=os.path.join(indir, 'grid.t001.nc')
RC=netcdf_tools.read_netcdf(gridfile,'RC')
```
