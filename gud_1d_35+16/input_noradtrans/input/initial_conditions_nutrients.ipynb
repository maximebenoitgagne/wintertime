# Generate plots of the initial conditions of the nutrients

```python
indir='../../../../output/run_20200113_0001_green_edge_10y/'
```

```python
import matplotlib
import matplotlib.pyplot as plt
import netcdf_tools
import numpy as np
import os.path
```

```python
gridfile=os.path.join(indir, 'grid.t001.nc')
RCfull=netcdf_tools.read_netcdf(gridfile,'RC')
bathymetry=-373.18
nbdepth=sum(RCfull>bathymetry)
RC=RCfull[0:nbdepth]
```

```python
# adapted from
# https://stackoverflow.com/questions/3899980/how-to-change-the-font-size-on-a-matplotlib-plot
# answer of Pedro M Duarte
SMALL_SIZE = 8
MEDIUM_SIZE = 12
BIGGER_SIZE = 14

# plt.rc('font', size=MEDIUM_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=BIGGER_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=MEDIUM_SIZE)    # fontsize of the x tick labels
plt.rc('ytick', labelsize=MEDIUM_SIZE)    # fontsize of the y tick labels
# plt.rc('legend', fontsize=MEDIUM_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
```

## $NO_3$

```python
infileNO3='1D_BB_NO3_GE_spring.delZ1016mm.32bits.bin'
inNO3full=np.fromfile(infileNO3, dtype='>f')
inNO3=inNO3full[0:nbdepth]
plt.figure(figsize=(3,5))
plt.plot(inNO3,RC)
plt.title('$NO_3$')
plt.xlabel('$NO_3\ (mmol\ N\ m^{-3})$')
plt.ylabel('$Depth\ (m)$')
plt.xlim(0,20)
plt.ylim(-400,0)
plt.grid(True)
plt.tight_layout()
plt.savefig('1D_BB_NO3_GE_spring.delZ1016mm.32bits.png')
```

## $PO_4$

```python
infilePO4='1D_BB_PO4_GE_spring.delZ1016mm.32bits.bin'
inPO4full=np.fromfile(infilePO4, dtype='>f')
inPO4=inPO4full[0:nbdepth]
plt.figure(figsize=(3,5))
plt.plot(inPO4,RC)
plt.title('$PO_4$')
plt.xlabel('$PO_4\ (mmol\ P\ m^{-3})$')
locs, labels=plt.yticks()
plt.yticks(locs, '')
plt.xlim(0,2)
plt.ylim(-400,0)
plt.grid(True)
plt.tight_layout()
plt.savefig('1D_BB_PO4_GE_spring.delZ1016mm.32bits.png')
```

## $SiO_2$

```python
infileSiO2='1D_BB_Si_GE_spring.delZ1016mm.32bits.bin'
inSiO2full=np.fromfile(infileSiO2, dtype='>f')
inSiO2=inSiO2full[0:nbdepth]
plt.figure(figsize=(3,5))
plt.plot(inSiO2,RC)
plt.title('$SiO_2$')
plt.xlabel('$SiO_2\ (mmol\ Si\ m^{-3})$')
locs, labels=plt.yticks()
plt.yticks(locs, '')
plt.xlim(0,30)
plt.ylim(-400,0)
plt.grid(True)
plt.tight_layout()
plt.savefig('1D_BB_Si_GE_spring.delZ1016mm.32bits.png')
```
